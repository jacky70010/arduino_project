```python
#先pip install pyserial
import serial  # 引用pySerial模組
import pygame
#pygame
pygame.mixer.init() 
track = pygame.mixer.music.load("Jonas Blue - Fast Car ft. Dakota.mp3")
COM_PORT = '/dev/ttyACM5'    # 指定通訊埠名稱
BAUD_RATES = 9600    # 設定傳輸速率
ser = serial.Serial(COM_PORT, BAUD_RATES)   
i=0
pygame.mixer.music.play()
try:
    while True:
        while ser.in_waiting:          # 若收到序列資料…
            if i==0:
                pygame.mixer.music.play()
                i+=1
            data_raw = ser.readline() 
            data = data_raw.decode()   # 用預設的UTF-8解碼
            if data[0:-2]=="press":
                pygame.mixer.music.pause()  
                print('接收到的資料：', data)
            if data[0:-2]=="hold":
                pygame.mixer.music.unpause()   
                print('////接收到的資料：', data)
            #print('接收到的原始資料：', data_raw)
            #print('接收到的資料：', data)
            #print('接收到的資料：', len(data))
            #print('接收到的資料：', data[0:-1])

except KeyboardInterrupt:
    pygame.mixer.music.pause()  
    ser.close()    # 清除序列通訊物件
    print('再見！')
```
