```python
#conda install tk 
#pip install pillow

import tkinter as tk
from functools import partial
from PIL import Image
import pygame

pygame.mixer.init() 
print("play it ")

#pygame.mixer.music.play()

#window initialization
window = tk.Tk()
window.title('my window')
window.geometry('1400x900')


#添加畫布
var = tk.StringVar()
canvas = tk.Canvas(window,  
    width = 1400,      # 指定Canvas组件的宽度  
    height = 900,      # 指定Canvas组件的高度  
    bg = 'white')      # 指定Canvas组件的背景色  
#im = Tkinter.PhotoImage(file='img.gif')     # 使用PhotoImage打开图片  
img = Image.open("head.png") 
img = img.resize((150, 750))
im = tk.PhotoImage(file='head.png')

canvas.create_image(700,750,image = im, anchor="center")      # 使用create_image将图片添加到Canvas组件中  
#canvas.create_text(302,77,       # 使用create_text方法在坐标（302，77）处绘制文字  
 #  text = 'Use Canvas'      # 所绘制文字的内容  
 #  ,fill = 'gray')       # 所绘制文字的颜色为灰色  
def on_change(varname, index, mode):
    canvas.itemconfigure(txt, text=window.getvar(varname))
    
txt = canvas.create_text(650,500, font='helvetica 24 bold italic', text=var.get(), fill = 'white')
var.trace_variable('w', on_change)



canvas.pack()         # 将Canvas添加到主窗口  



#l = tk.Label(window, textvariable=var, bg='white', font=('Arial', 12), width=100,       height=30)
#l = tk.Label(window, text='OMG! this is TK!', bg='green', font=('Arial', 12), width=15, height=2)
#l.place(x=300, y=250)

ans1="a"
ans2="b"
ans3="c"
on_hit = False
def press(a):
    global on_hit
    print(a)
    if a==ans1:
        track = pygame.mixer.music.load("applause.mp3")
        pygame.mixer.music.play()
        var.set("消化系統有那些？")
    if a==ans2:
        track = pygame.mixer.music.load("wronganswer.mp3")
        pygame.mixer.music.play()
        var.set("呼吸系統有那些？")
    if a==ans3:
        var.set("排泄系統有那些？")
def submit():
    print("finish")
    
    var.set("finish")
b = tk.Button(window, bg='red', text='submit', width=30, height=10, command=submit)
#b.pack()
b.place(x=1150, y=750)



button_img_gif = tk.PhotoImage(file = '2.gif')
button_img = tk.Button(window, image = button_img_gif, text = '11', width=150, height=200, command=partial(press, ans1))
button_img.place(x=100, y=20)

button_img_gif2 = tk.PhotoImage(file = '2.gif')
button_img2 = tk.Button(window, image = button_img_gif, text = '11', width=150, height=200, command=partial(press, ans2))
button_img2.place(x=350, y=20)

button_img2 = tk.Button(window, image = button_img_gif, text = '11', width=150, height=200, command=partial(press, ans3))
button_img2.place(x=600, y=20)

button_img2 = tk.Button(window, image = button_img_gif, text = '11', width=150, height=200, command=partial(press, ans3))
button_img2.place(x=850, y=20)

button_img2 = tk.Button(window, image = button_img_gif, text = '11', width=150, height=200, command=partial(press, ans3))
button_img2.place(x=1100, y=20)

window.mainloop()
```

```python
import tkinter         # 导入Tkinter模块  
from PIL import Image, ImageTk  
  
root = tkinter.Tk()  
canvas = tkinter.Canvas(root,  
    width = 500,      # 指定Canvas组件的宽度  
    height = 600,      # 指定Canvas组件的高度  
    bg = 'white')      # 指定Canvas组件的背景色  
#im = Tkinter.PhotoImage(file='img.gif')     # 使用PhotoImage打开图片  
image = Image.open("11.gif")  
im = ImageTk.PhotoImage(image)  
  
canvas.create_image(200,200,image = im)      # 使用create_image将图片添加到Canvas组件中  
canvas.create_text(302,77,       # 使用create_text方法在坐标（302，77）处绘制文字  
   text = 'Use Canvas'      # 所绘制文字的内容  
   ,fill = 'gray')       # 所绘制文字的颜色为灰色  
canvas.create_text(300,75,  
   text = 'Use Canvas',  
   fill = 'blue')  
canvas.pack()         # 将Canvas添加到主窗口  
root.mainloop()  
```
